```python
import pandas as pd 
file_path='eCommerce Store.xlsx'
df=pd.read_excel(file_path)
#Display the first few rows of the dataframe 
print(df.head())
print(df.tail())
```

```python
#summary statistics for numerical columns
print(df.describe())

#print columns names to check for 'channel'
print(df.columns)
```

```python
print(df.info)
```

```python
#clean the gender column
df['Gender']=df['Gender'].replace({'W':'Women','M':'Men'})
print(df)
                                   
```

```python
#clean the to QTY column
df['Qty']=df['Qty'].replace({'One':1,'Two':2})
print(df)
```

```python
#convet city and state columns to uppercase
df['ship-city']=df['ship-city'].str.upper()
df['ship-state']=df['ship-state'].str.upper()
print(df)
```

```python
def categorize_age(age):
    if age>60:
        return 'Above 60'
    elif 20 <=age<= 60:
        return '20 to 60'
    else:
        return 'Below 20'
df['Age Group']=df['Age'].apply(categorize_age)
print(df)
```

```python
df['Date']=pd.to_datetime(df['Date'],format='%d-%m-%Y')
df['Month']=df['Date'].dt.month_name()
print(df['Month'])
```

```python
df['Business Type']=df['B2B'].replace({False:'B2C',True:'B2B'})
print(df['Business Type'].to_string())
```

```python
import  pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.barplot(x=df['Gender'].value_counts().index,y=df.groupby('Gender')['Amount'].sum().values,hue=df['Gender'].value_counts().index,palette='pastel')
plt.title('Total Sale by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sale (INR)')
plt.subplot(1,2,2)
sns.barplot(x=df['Gender'].value_counts().index,y=df['Gender'].value_counts().values,hue=df['Gender'].value_counts().index,palette='pastel')
plt.title('Total Order by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Order')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(data=df,x='Status',hue='Status',palette='viridis')
plt.title('Order Status Distribution ')
plt.xlabel('Order Status')
plt.ylabel('Count')
plt.show()
```

```python
state_sales=df.groupby('ship-state')['Amount'].sum().reset_index()
state_sales_sorted=state_sales.sort_values(by='Amount',ascending=False)
top_states=state_sales_sorted.head(5)
plt.figure(figsize=(10,6))
ax=sns.barplot(x='Amount',y='ship-state',data=top_states)
plt.title('Top 5 States by Sales')
plt.xlabel('Total Sales(INR)')
plt.ylabel('Sate')
plt.show()
```

```python
city_sales=df.groupby('ship-city')['Amount'].sum().reset_index()
city_sales_sorted=city_sales.sort_values(by='Amount',ascending=False)
top_city=city_sales_sorted.head(5)
plt.figure(figsize=(10,6))
ax=sns.barplot(x='Amount',y='ship-city',data=top_city,palette='magma',hue='ship-city')
plt.title('Top 5 city by Sales')
plt.xlabel('Total Sales(INR)')
plt.ylabel('Sate')
plt.show()
```

```python
#plotting
plt.figure(figsize=(12,8))
sns.scatterplot(data=df,x='Age',y='Amount',hue='Gender',palette='dark',alpha=0.7)
plt.title('Relationship Between Age and Sales Amount by Gender')
plt.xlabel('Age')
plt.ylabel('Sales Amount (INR)')
plt.legend(title='Gender')
plt.show()
```

```python
channel_sales=df.groupby('Channel ')['Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=channel_sales.index,y=channel_sales.values,palette='cubehelix',hue=channel_sales.index)
plt.title(' Sales Contribution by Channel')
plt.xlabel('Channel')
plt.ylabel('Total Sales (INR)')
plt.xticks(rotation=45)
plt.show()
```

```python
category_sales=df.groupby('Category')['Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=category_sales.index,y=category_sales.values,palette='crest',hue=category_sales.index)
plt.title(' Sales Contribution by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales (INR)')
plt.xticks(rotation=45)
plt.show()
```

```python
#caculate total sale by business type
b2b_c2c_sales=df.groupby('Business Type')['Amount'].sum()
plt.figure(figsize=(8,6))
sns.barplot(x=b2b_c2c_sales.index,y=b2b_c2c_sales.values,palette='Set2',hue=b2b_c2c_sales.index)
plt.title('B2B vs B2C Transactions')
plt.xlabel('Business Type')
plt.ylabel('Total Sales (INR)')
plt.show()
```
